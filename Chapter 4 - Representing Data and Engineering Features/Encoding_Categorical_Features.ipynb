## One-Hot Encoding (Dummy Variables)

```
# Using pandas instead of Scikit-learn

import pandas as pd
data = pd.read_csv("adult.data", header=None, index_col=False,
                   names =['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                           'marital-status', 'occupation', 'relationship', 'race', 'gender',
                           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
                           'income'])

# For practive purposes we only use some of the features
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]

# I.Python display allows nice output formatting
display(data.head())
```

### Checking string - encoded Categorical Values

```
# Checking for meaningful - categorical data. There might be differences in which the data is entered by different users. Gender = Male
# Man are similar and thus we should know the number of values in a particular column feature

print(data.gender.value_counts())
```

```
# When working on a real - project you should always check if the data in the dataset is represented in a correct format. Here we skip
# it because I know it is in the correct format

# To encode this categorical data we use the get_dummies function which is from the Pandas library, and it automatically transforms the
# cateegorical or object type (like strings)

print("Original Features:\n", list(data.columns), "\n")
data_dummies = pd.get_dummies(data)
print("Feature after get dummies:\n", list(data_dummies.columns))
```

```
data_dummies.head()
```

```
# We will now use the 'values' attribute to convert the data_dummies DataFrame into a NumPy Array, and then train a machine learning
# model on it.

# We have to seperate the target variable (which is now encoded in 2 income columns) from the data before training the model.

# Including such an output variable or some derived output variable into the feature representation is a very common mistake in building
# supervised machine learning models.

# Indexing of Pandas DataFrame and a NumPy array are different.

features = data_dummies.iloc[:, 0:-3]

# Extracting the NumPy Arrays
X = features.values
y = data_dummies['income_ >50K'].values
print("X.shape: {} y.shape: {}".format(X.shape, y.shape))
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
print("Test score: {:.2f}".format(logreg.score(X_test, y_test)))
```

## Numbers can Encode Categoricals

```
# When numbers/integers are used for categorical data we cannot use 'get_dummies' function of the Pandas Library. To get around this we
# use the OneHotEncoder of the scikit learn library. 

# It allows us to set which variables are dicrete and which are continous. Or they conver the numeric values of the dataset into strings.
```

```
# Creating a DataFrame with an integer feature and a categorical string feature

demo_df = pd.DataFrame({'Integer Feature': [0, 1, 2, 1],
                        'Categorical Feature': ['socks', 'fox', 'socks', 'box']})

display(demo_df)
```

```
# Encoding using the get_dummies will only change the string feature and not the integer feature as you can see below

pd.get_dummies(demo_df)
```

```
# You need to explicitly change the type of the integer feature as string to use 'get_dummies' function on it.

demo_df['Integer Feature'] = demo_df['Integer Feature'].astype(str)
pd.get_dummies(demo_df, columns=['Integer Feature', 'Categorical Feature'])
```
